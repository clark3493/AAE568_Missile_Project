# VARIOUS ASSUMPTIONS TO REMEMBER FOR REPORT

- Likely will not be including changes in air density with respect to altitude, but may be feasible
- Not including aeroelastic effects
- Limit to linear aerodynamic region
    - Book makes a good point that if wind is includes, AoA's may be near 90 deg at launch... we will probably just want to not consider this
    - Book also considers that aerodynamic characteristics vary with Mach number... good realistic consideration, maybe not necessary for our application
- Assume effect of control surface deflection can be modeled as a rotation of a flat plate with a fixed point of lift and drag concentration
- No wake effects from leading control wings/control surfaces on trailing wings/control surfaces
- Constant mass (maybe not a great assumption is we include a propulsion system)
- Book [1] includes aerodynamic symmetry in roll (p38 in pdf)... not sure this is a great assumption.... probably doesn't matter for 2D problem though
- Multiple other assumptions listed on p38 of [1]

# COORDINATE SYSTEMS

Coordinate systems and coordinate are per Section 2.1 of [1]. It may become necessary to incorporate the quaternion system to avoid singularities when Euler angles approach $90^\circ$, as will happen when a near normal impact angle is desired.
TBD on if we use the Y-Z as shown in lower=left in above picture... may be easier to implement if we orient the fins with the axes.

Fixed-earth (inertial) axes: $$<X_E, Y_E, Z_E>$$
Body Axes: $$<X_B, Y_b, Z_B>$$
Euler Angles: $$<\phi, \theta, \psi>$$
where $\phi, \theta, \psi$ represent the Euler angles of the body axes with respect to the fixed earth axes. In our context, it is safe to assume that all the distances are relatively small with repsect to the radius of earth and therefore fixed-earth and inertial axis systems can be assumed to be approximately the same. This would likely not be a safe assumption for long range ballistic missiles.

The initial position of the missile and target in the fixed-earth axis system, as well as the missile initial attitude (i.e. $\phi, \theta, \psi$) will need to be set for each simulation run. The relative position of the missile to the target can then be used as a state, and the missiles attitude cna be used for body force calculations.
